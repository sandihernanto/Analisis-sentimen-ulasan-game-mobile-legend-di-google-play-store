```python
from google_play_scraper import app
import pandas as pd
import numpy as np
```

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.mobile.legends',
    lang='id',
    country='id',
    sort=Sort.NEWEST,
    count=500,
    filter_score_with=None
)
```

```python
df_busu = pd.DataFrame(np.array(result),columns=['review'])

df_busu = df_busu.join(pd.DataFrame(df_busu.pop('review').tolist()))

df_busu.head()
```

```python
len(df_busu.index) #kemudian hitung kembali berapa jumlah data yg didapatkan
```

```python
df_busu[['reviewId', 'userName', 'userImage', 'score','at', 'content']].head()
```

```python
new_df = df_busu[['reviewId', 'userName', 'userImage', 'score','at', 'content']]
sorted_df = new_df.sort_values(by='at', ascending=False) #Sort by Newst, change to True if you want to sort by Oldest.
sorted_df.head()
```

```python
my_df = sorted_df[['score', 'content']] #get score and content only
```
